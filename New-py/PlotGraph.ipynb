# Bubble chart

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
year =[2012,2013,2014,2015,2016,2017]
salary = [12,13,14,17,19,20]
population = [100,120,180,250,300,370]
c = np.arange(len(year))
```

```python
plt.scatter(year,salary,s=population , c =c , alpha=1,edgecolor='b')
plt.show()
```

```python
plt.scatter(year,salary,s=population ,marker ='^', c =c , alpha=1,edgecolor='b')
plt.text(year[0]+0.1,salary[0]+0.1,population[0])
plt.show()
```

```python
plt.scatter(year,salary,s=population ,marker ='^', c =c , alpha=1,edgecolor='b')
for i in range(len(year)):
    plt.text(year[i]+0.05,salary[i]+0.05,population[i])
plt.show()
```

# Pie chart

```python
sizes = [3,4,6,2]
color = ["blue" ,"purple","red","pink"]
label = ['A','B','C','D']
explode = [0.1,0.2,0,0]
explode = [0.1,0,0,0] # To highlight 1 particular piece
plt.pie(sizes,colors = color,labels = label,autopct = "%.1f%%",explode=explode,counterclock=False,startangle=100)
plt.title("split amoung classes")
plt.show()
```
